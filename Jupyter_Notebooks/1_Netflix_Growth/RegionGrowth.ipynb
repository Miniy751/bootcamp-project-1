# Exploring Subcriber and Revenue Growth in Each Region

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import file and convert revnue to billion"
file = "Resources/Revenue_Data.csv"
area_rev = pd.read_csv(file, index_col=0).reset_index()
area_rev['Revenue'] = area_rev['Revenue'].div(1000000000).round(2)
area_rev.head()
```

```python
#group by years and reorder labels
group_years_rev = area_rev.groupby("Years")
group_rev_df = group_years_rev["Revenue"].sum().reindex(["Q1 - 2018", "Q2 - 2018", "Q3 - 2018","Q4 - 2018",
                                                      "Q1 - 2019", "Q2 - 2019", "Q3 - 2019","Q4 - 2019",
                                                      "Q1 - 2020", "Q2 - 2020"]).reset_index()
#show dataframe to check numbers
group_rev_df.style.format({"Revenue":"${:,.2f}"})
```

```python
#assign values
x_axis = group_rev_df["Years"]
y_axis = group_rev_df["Revenue"]

#plot graph
plt.plot(x_axis, y_axis, marker ="o", label="")
plt.xticks(rotation=90)
plt.title("Netflix Global Revenue")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
```

```python
#get values of x and y axis for each region
x_usc = area_rev["Years"][area_rev["Area"] == "United States and Canada"]
x_ema = area_rev["Years"][area_rev["Area"] == "Europe,  Middle East and Africa"]
x_la = area_rev["Years"][area_rev["Area"] == "Latin America"]
x_ap = area_rev["Years"][area_rev["Area"] == "Asia-Pacific"]

y_usc = area_rev["Revenue"][area_rev["Area"] == "United States and Canada"]
y_ema = area_rev["Revenue"][area_rev["Area"] == "Europe,  Middle East and Africa"]
y_la = area_rev["Revenue"][area_rev["Area"] == "Latin America"]
y_ap = area_rev["Revenue"][area_rev["Area"] == "Asia-Pacific"]
```

```python
# plotting the line 1 points 
plt.plot(x_usc, y_usc, label = "United States and Canada", marker ="o")
plt.plot(x_ema, y_ema, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(x_la, y_la, label = "Latin America", marker ="o")
plt.plot(x_ap, y_ap, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Revenue By Region")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```

```python
file = "Resources/Subscriber_Data.csv"
area_sub = pd.read_csv(file, index_col=0).reset_index()
area_sub['Subscribers'] = area_sub['Subscribers'].div(1000000).round(2)
area_sub.head()
```

```python
#group by years and reorder labels
group_years_sub = area_sub.groupby("Years")
group_sub_df = group_years_sub["Subscribers"].sum().reindex(["Q1 - 2018", "Q2 - 2018", "Q3 - 2018","Q4 - 2018",
                                                      "Q1 - 2019", "Q2 - 2019", "Q3 - 2019","Q4 - 2019",
                                                      "Q1 - 2020", "Q2 - 2020"]).reset_index()

#show dataframe to check numbers
group_sub_df
```

```python
#assign values
x_axis = group_sub_df["Years"]
y_axis = group_sub_df["Subscribers"]

#plot graph
plt.plot(x_axis, y_axis, marker ="o", label="")
plt.xticks(rotation=90)
plt.title("Netflix Global Subcribers")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Subscribers (Million)")
plt.tight_layout()
```

```python
#get values of x and y axis for each region
x_usc = area_sub["Years"][area_sub["Area"] == "United States and Canada"]
x_ema = area_sub["Years"][area_sub["Area"] == "Europe,  Middle East and Africa"]
x_la = area_sub["Years"][area_sub["Area"] == "Latin America"]
x_ap = area_sub["Years"][area_sub["Area"] == "Asia-Pacific"]

y_usc = area_sub["Subscribers"][area_sub["Area"] == "United States and Canada"]
y_ema = area_sub["Subscribers"][area_sub["Area"] == "Europe,  Middle East and Africa"]
y_la = area_sub["Subscribers"][area_sub["Area"] == "Latin America"]
y_ap = area_sub["Subscribers"][area_sub["Area"] == "Asia-Pacific"]
```

```python
# plotting the line 1 points
plt.plot(x_usc, y_usc, label = "United States and Canada", marker ="o")
plt.plot(x_ema, y_ema, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(x_la, y_la, label = "Latin America", marker ="o")
plt.plot(x_ap, y_ap, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Subscribers By Region")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Subscribers (Million)")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```

```python
years = ["Q1 - 2018", "Q2 - 2018", "Q3 - 2018","Q4 - 2018",
           "Q1 - 2019", "Q2 - 2019", "Q3 - 2019","Q4 - 2019",
           "Q1 - 2020", "Q2 - 2020"]

y_sub = group_sub_df["Subscribers"].pct_change()
y_rev = group_rev_df["Revenue"].pct_change()

plt.plot(years, y_sub, marker ="o", label="Subscriber")
plt.plot(years, y_rev, marker ="o", label="Revenue")
plt.xticks(rotation=90)
plt.title("Global Revenue and Subscriber Growth")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Subscribers (Million)")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```

```python
y_usc_sub = area_sub["Subscribers"][area_sub["Area"] == "United States and Canada"].pct_change()
y_ema_sub = area_sub["Subscribers"][area_sub["Area"] == "Europe,  Middle East and Africa"].pct_change()
y_la_sub = area_sub["Subscribers"][area_sub["Area"] == "Latin America"].pct_change()
y_ap_sub = area_sub["Subscribers"][area_sub["Area"] == "Asia-Pacific"].pct_change()

y_usc_rev = area_rev["Revenue"][area_rev["Area"] == "United States and Canada"].pct_change()
y_ema_rev = area_rev["Revenue"][area_rev["Area"] == "Europe,  Middle East and Africa"].pct_change()
y_la_rev = area_rev["Revenue"][area_rev["Area"] == "Latin America"].pct_change()
y_ap_rev = area_rev["Revenue"][area_rev["Area"] == "Asia-Pacific"].pct_change()
```

```python
# plotting the line 1 points
plt.plot(years, y_usc_sub, label = "United States and Canada", marker ="o")
plt.plot(years, y_ema_sub, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(years, y_la_sub, label = "Latin America", marker ="o")
plt.plot(years, y_ap_sub, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Subscriber Growth")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Percentage Change")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```

```python
# plotting the line 1 points
plt.plot(years, y_usc_rev, label = "United States and Canada", marker ="o")
plt.plot(years, y_ema_rev, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(years, y_la_rev, label = "Latin America", marker ="o")
plt.plot(years, y_ap_rev, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Percent Revenue Growth")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Percentage Change")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```

```python
plt.plot(years, y_usc_sub, label = "Subscribers", marker ="o")
plt.plot(years, y_usc_rev, label = "Revenue", marker ="o")
plt.xticks(rotation=90)
plt.title("United States and Canada Growth")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Percentage Change")
plt.tight_layout()
plt.legend(loc="best")
```

```python
figure, axis = plt.subplots(2, 2)

# US, CN
axis[0, 0].plot(years, y_usc_sub, marker ="o")
axis[0, 0].plot(years, y_usc_rev, marker ="o")
axis[0, 0].set_title("United States and Canada")
axis[0, 0].tick_params(axis='x', rotation=45)
# EU, ME, AF
axis[0, 1].plot(years, y_ema_sub, marker ="o")
axis[0, 1].plot(years, y_ema_rev, marker ="o")
axis[0, 1].set_title("Europe, Middle East and Africa")
axis[0, 1].tick_params(axis='x', rotation=45)
# AP
axis[1, 0].plot(years, y_ap_sub, marker ="o")
axis[1, 0].plot(years, y_ap_rev, marker ="o")
axis[1, 0].set_title("Asia Pacific")
axis[1, 0].tick_params(axis='x', rotation=45)
# LA
axis[1, 1].plot(years, y_la_sub, marker ="o",label = "Subscribers")
axis[1, 1].plot(years, y_la_rev, marker ="o",label = "Revenue")
axis[1, 1].set_title("Latin America")
axis[1, 1].tick_params(axis='x', rotation=45)

plt.legend(bbox_to_anchor=(1.8,1.6), borderaxespad=0)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=1)
```
