```python
#Import dependecies 
import pandas as pd
import matplotlib as plt
```

```python
#Read Csv file into a DataFrame 
file = pd.read_csv("Resources/clean_netflix_shows.csv")

file.head()
```

```python
year_count = file["Year"].value_counts()

df= pd.DataFrame(year_count)

df
```

```python
grouped_yearcount = file.groupby("Year")

print(grouped_yearcount)

grouped_count = grouped_yearcount.size()

groupby_yearcount_df = pd.DataFrame(grouped_count)

groupby_yearcount_df
```

```python
average_year_runtime = file.groupby("Year")["Runtime"].mean()

average_year_runtime_df = pd.DataFrame(average_year_runtime)

average_year_runtime_df
```

```python
genre_count = file["Genre"].value_counts()

genre_count_df = pd.DataFrame(genre_count)

genre_count_df
```

```python
grouped_genre = file.groupby("Genre")

print(grouped_genre)

grouped_genrecount = grouped_genre.size()

grouped_genrecount_df = pd.DataFrame(grouped_genrecount)

grouped_genrecount_df
```

```python
average_genre_runtime = file.groupby("Genre")["Runtime"].mean()

average_genre_runtime_df = pd.DataFrame(average_genre_runtime)

average_genre_runtime_df
```

```python
runtime_min = file["Runtime"].min()

runtime_min
```

```python
runtime_max = file["Runtime"].max()

runtime_max
```

```python
runtime_median = file["Runtime"].median()

runtime_median
```

```python
runtime_bins = [0,96.0,97.0,209]

categories = ["short", "Medium", "Long"]
```

```python
file["Runtime"] = pd.cut(file["Runtime"], runtime_bins, labels = categories, include_lowest = True)

file
```
