# Exploring Subcriber and Revenue Growth in Each Region

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import file and convert revnue to billion"
file = "Resources/Revenue_Data.csv"
area_rev = pd.read_csv(file, index_col=0).reset_index()
area_rev['Revenue'] = area_rev['Revenue'].div(1000000000).round(2)
area_rev.head()
```

```python
#group by years and reorder labels
group_years_rev = area_rev.groupby("Years")
group_rev_df = group_years_rev["Revenue"].sum().reindex(["Q1 - 2018", "Q2 - 2018", "Q3 - 2018","Q4 - 2018",
                                                      "Q1 - 2019", "Q2 - 2019", "Q3 - 2019","Q4 - 2019",
                                                      "Q1 - 2020", "Q2 - 2020"]).reset_index()
#show dataframe to check numbers
group_rev_df.style.format({"Revenue":"${:,.2f}"})
```

```python
#assign values
x_axis = group_rev_df["Years"]
y_axis = group_rev_df["Revenue"]

#plot graph
plt.plot(x_axis, y_axis, marker ="o", label="")
plt.xticks(rotation=90)
plt.title("Netflix Global Revenue")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
```

```python
#get values of x and y axis for each region
x_usc = area_rev["Years"][area_rev["Area"] == "United States and Canada"]
x_ema = area_rev["Years"][area_rev["Area"] == "Europe,  Middle East and Africa"]
x_la = area_rev["Years"][area_rev["Area"] == "Latin America"]
x_ap = area_rev["Years"][area_rev["Area"] == "Asia-Pacific"]

y_usc = area_rev["Revenue"][area_rev["Area"] == "United States and Canada"]
y_ema = area_rev["Revenue"][area_rev["Area"] == "Europe,  Middle East and Africa"]
y_la = area_rev["Revenue"][area_rev["Area"] == "Latin America"]
y_ap = area_rev["Revenue"][area_rev["Area"] == "Asia-Pacific"]
```

```python
# plotting the line 1 points 
plt.plot(x_usc, y_usc, label = "United States and Canada", marker ="o")
plt.plot(x_ema, y_ema, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(x_la, y_la, label = "Latin America", marker ="o")
plt.plot(x_ap, y_ap, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Revenue By Region")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```
