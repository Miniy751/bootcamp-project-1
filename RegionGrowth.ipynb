# Exploring Subcriber and Revenue Growth in Each Region

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#import file and convert revnue to billion"
file = "Resources/Revenue_Data.csv"
netflix_omdb_df = pd.read_csv(file, index_col=0).reset_index()
netflix_omdb_df['Revenue'] = netflix_omdb_df['Revenue'].div(1000000000).round(2)
netflix_omdb_df
```

```python
#group by years and reorder labels
group_years = netflix_omdb_df.groupby("Years")
group_year_df = group_years["Revenue"].sum().reindex(["Q1 - 2018", "Q2 - 2018", "Q3 - 2018","Q4 - 2018",
                                                      "Q1 - 2019", "Q2 - 2019", "Q3 - 2019","Q4 - 2019",
                                                      "Q1 - 2020", "Q2 - 2020"]).reset_index()
#show dataframe to check numbers
group_year_df.style.format({"Revenue":"${:,.2f}"})
```

```python
#assign values
x_axis = group_year_df["Years"]
y_axis = group_year_df["Revenue"]

#plot graph
plt.plot(x_axis, y_axis, marker ="o", label="")
plt.xticks(rotation=90)
plt.title("Netflix Global Revenue By Financial Quarter")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
```

```python
#get values of x and y axis for each region
x_usc = netflix_omdb_df["Years"][netflix_omdb_df["Area"] == "United States and Canada"]
x_ema = netflix_omdb_df["Years"][netflix_omdb_df["Area"] == "Europe,  Middle East and Africa"]
x_la = netflix_omdb_df["Years"][netflix_omdb_df["Area"] == "Latin America"]
x_ap = netflix_omdb_df["Years"][netflix_omdb_df["Area"] == "Asia-Pacific"]

y_usc = netflix_omdb_df["Revenue"][netflix_omdb_df["Area"] == "United States and Canada"]
y_ema = netflix_omdb_df["Revenue"][netflix_omdb_df["Area"] == "Europe,  Middle East and Africa"]
y_la = netflix_omdb_df["Revenue"][netflix_omdb_df["Area"] == "Latin America"]
y_ap = netflix_omdb_df["Revenue"][netflix_omdb_df["Area"] == "Asia-Pacific"]
```

```python
# plotting the line 1 points 
plt.plot(x_usc, y_usc, label = "United States and Canada", marker ="o")
plt.plot(x_ema, y_ema, label = "Europe, Middle East and Africa", marker ="o")
plt.plot(x_la, y_la, label = "Latin America", marker ="o")
plt.plot(x_ap, y_ap, label = "Asia-Pacific", marker ="o")

plt.xticks(rotation=90)
plt.title("Revenue By Region")
plt.xlabel("Financial Quarter - Year")
plt.ylabel("Revenue (Billion US Dollars)")
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```
