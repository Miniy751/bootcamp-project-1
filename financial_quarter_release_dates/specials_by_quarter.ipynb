```python
import pandas as pd
import numpy as np
```

```python
file = "../Resources/omdb_pull"
omdb_pull = pd.read_csv(file, index_col=0)
omdb_pull
netflix_omdb_df = omdb_pull
```

```python
#return month of each premiere
netflix_omdb_df["Month"] = netflix_omdb_df["Premiere"].str.split().str[0].str.strip()
netflix_omdb_df["Year"] = netflix_omdb_df["Premiere"].str.split().str[-1].str.strip()

#list financial quarters to sort to
q1 = ["January", "February", "March"]  
q2 = ["April","May","June"]
q3 = ["July","August","September"]  
q4 = ["October", "November", "December"]

#Add New Column
netflix_omdb_df["Financial Quarter"] = np.nan

for index, row in netflix_omdb_df.iterrows():
    key = row["Month"]
    if key in q1:
        netflix_omdb_df.loc[index, "Financial Quarter"] = "1"
    elif key in q2:
        netflix_omdb_df.loc[index, "Financial Quarter"] = "2"
    elif key in q3:
        netflix_omdb_df.loc[index, "Financial Quarter"] = "3"
    elif key in q4:
        netflix_omdb_df.loc[index, "Financial Quarter"] = "4"
        
netflix_omdb_df
```

```python
netflix_omdb_df = netflix_omdb_df[netflix_omdb_df.isna().any(axis=1)]
netflix_omdb_df["Financial Quarter"].value_counts()
netflix_omdb_df
```

```python
netflix_omdb_df.to_csv("../Resources/netflix_omdb")
```
